```python
nums = [1,3,-1,-3,5,3,6,7]
k = 3
```

```python
def sol(nums, k):
    
    l = 0
    dq = deque()
    res = []

    for r in range(len(nums)):

        while dq and nums[r] > dq[-1]:
            dq.pop()
        dq.append(nums[r])

        if r > k - 2:
            while dq[0] < r - k + 1:
                dq.popleft()
            res.append(dq[0])
            l += 1

    return res
    
```

```python
sol(nums, k)
```
