```python
nums = [1,2,1,2,1,2,3,1,3,2]
k = 2
```

```python
def sol(nums, k):
    d = dict()

    for i in nums:
        if i not in d:
            d[i] = [i, 0]
        d[i][1] += 1

    print(d)

    lst = list(d.values())

    lst_sorted = sorted(lst, key=lambda row: row[1], reverse=True)

    res = []

    for i in lst_sorted[:k]:
        res.append(i[0])

    return res
```

```python
sol(nums, k)
```

```python
def sol2(nums, k):
    if len(nums) == k:
        return nums
    
    d = dict()

    for i in nums:
        d[i] = d.get(i,0) + 1

    print(d)

    buckets = []

    for i in range(len(nums) + 1):
        buckets.append([])

    for key, value in d.items():
        buckets[value].append(key)

    print(buckets)

    res = []

    for i in range(len(buckets) - 1, 0, -1):
        for n in buckets[i]:
            res.append(n)
            if len(res) == k:
                return res
    


    
```

```python
sol2(nums, k)
```

```python
def sol3(nums, k):
    if len(nums) == k:
        return nums
    
    d = dict()

    for i in nums:
        d[i] = d.get(i,0) + 1

    return heapq.nlargest(k, d.keys(), key=d.get)

    
```
