```python
from collections import Counter

s = "ADOBECODEBANC"
t = "ABC"
```

```python
def sol(s,t):

    if len(s) < len(t):
        return ''

    goal = {}
    for ch in t:
        goal[ch] = goal.get(ch, 0) + 1

    window = {}

    formed = 0
    required = len(goal)

    l = 0
    start = 0
    minLen = float('inf')
    

    for r, ch in enumerate(s):

        window[ch] = window.get(ch, 0) + 1
        if ch in goal and window[ch] == goal[ch]:
            formed += 1

        
        while formed == required:
            
            curr_len = r-l+1
            
            if curr_len < minLen:
                minLen = curr_len
                start = l

            # сначала убираем все даные по текущему l, только потом двигаем указатель
            left = s[l]
            window[left] -= 1
            if left in goal and window[left] < goal[left]:
                formed -= 1
            l += 1
            
            
    return '' if minLen == float('inf') else s[start:start+minLen]
        
```

```python
sol(s,t)
```
