```python
def sol(s,t):

    need = {}
    for ch in t:
        need[ch] = need.get(ch, 0) + 1

    window = {}
    formed = 0
    required = len(t)
    l = 0
    word = s


    for r in range(len(s)):
        
        window[s[r]] = window.get(s[r], 0) + 1
        
        if s[r] in need and window[s[r]] >= need[s[r]]:
            formed += 1

        print()
        print(s[l:r+1])
        print(f'formed: {formed} required:{required}')

        while formed == required:
            print('--- he-lo ! ðŸ¥³')

            if s[l] in need:
        
                print('--- he-lo ! ðŸ¥³')
        
    return word
        
    
```
