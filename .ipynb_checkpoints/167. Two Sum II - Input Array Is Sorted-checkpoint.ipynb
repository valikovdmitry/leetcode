```python
numbers = [-1,0]
target = -1
```

```python
def sol(numbers, target):
    left = 0
    right = len(numbers) - 1

    while left < right:
        curr = numbers[left] + numbers[right] 

        if curr < target:
            left += 1

        elif curr > target:
            right -= 1

        else:
            return [left + 1, right + 1]
    

    
            

        
```

```python
sol(numbers, target)
```
