```python
board = [
     ["5","3",".",".","7",".",".",".","."]
    ,["6",".",".","1","9","5",".",".","."]
    ,[".","9","4",".",".",".",".","6","."]
    ,["8",".",".",".","6",".",".",".","3"]
    ,["4",".",".","8",".","3",".",".","1"]
    ,["7",".",".",".","2",".",".",".","6"]
    ,[".","6",".",".",".",".","2","8","."]
    ,[".",".",".","4","1","9",".",".","5"]
    ,[".",".",".",".","8",".",".","7","9"]]
```

```python
board
```

```python
def sol4(board):
    
    columns = [set() for i in range(9)]
    blocks  = [[set() for i in range(3)] for i in range(3)]
    

    # прохожу по строкам
    for row_idx, row in enumerate(board):

        # создаю сет для проверки уникальности
        row_seen = set()

        # прохожу по значениям
        for val_idx, value in enumerate(row):

            # пропускаю точки, то есть пустые поля
            if value == '.':
                continue
            
            # проверяю строку на уникальность через сет 'd'
            if value in row_seen:
                return False
            row_seen.add(value)
            
            # проверяю колонку на уникальность через заранее заготовленный сет в массиве 'columns'
            if value in columns[val_idx]:
                return False
            columns[val_idx].add(value)

            # проверяю блок на уникальность через заранее заготовленный сет в двумерном массиве 'blocks'
            # я могу по row_idx = i и val_idx = j понимать в какой сет блоков попадает точка
            i = row_idx//3
            j =val_idx//3
            
            if value in blocks[i][j]:
                return False
            blocks[i][j].add(value)


    return True
```

```python
sol4(board)
```
