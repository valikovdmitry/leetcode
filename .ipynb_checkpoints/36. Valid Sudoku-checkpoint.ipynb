```python
def sol2(board):

    columns = [[], [], [], [], [], [], [], [], []]

    blocks_1 = [[], [], []]
    blocks_2 = [[], [], []]
    blocks_3 = [[], [], []]

    # прохожу по строкам
    for row_idx, row in enumerate(board):
        d = {}
        for idx, value in enumerate(row):
            # делаю счетчик строки
            d[value] = d.get(value, 0) + 1

            # собираю колонки мимо ходом
            columns[idx].append(value)

            # собираю блоки мимо ходом тоже
            if row_idx in [0,1,2]:
                if idx in [0,1,2]:
                    blocks_1[0].append(value)
                elif idx in [3,4,5]:
                    blocks_1[1].append(value)
                elif idx in [6,7,8]:
                    blocks_1[2].append(value)
            
            elif row_idx in [3,4,5]:
                if idx in [0,1,2]:
                    blocks_2[0].append(value)
                elif idx in [3,4,5]:
                    blocks_2[1].append(value)
                elif idx in [6,7,8]:
                    blocks_2[2].append(value)

            elif row_idx in [6,7,8]:
                if idx in [0,1,2]:
                    blocks_3[0].append(value)
                elif idx in [3,4,5]:
                    blocks_3[1].append(value)
                elif idx in [6,7,8]:
                    blocks_3[2].append(value)

                    
        # проверяю соответсвие правилу по строчкам
        for key, value in d.items():
            if value != 1 and key != '.':
                return False

    # for i in blocks_1:
    #     print(i)
    # for i in blocks_2:
    #     print(i)
    # for i in blocks_3:
    #     print(i)
        
    # print('выше это блоки')

    
    # прохожу по колонкам
    for column in columns:
        d2 = {}

        # делаю счетчик строки
        for idx, value in enumerate(column):
            # делаю счетчик строки
            d2[value] = d2.get(value, 0) + 1

        # проверяю соответсвие правилу по колонкам
        for key, value in d2.items():
            if value != 1 and key != '.':
                return False


    blocks = blocks_1 + blocks_2 + blocks_3
    
    # прохожу по блокам
    for block in blocks:
        d3 = {}
    
        # делаю счетчик блока
        for idx, value in enumerate(block):
            # делаю счетчик строки
            d3[value] = d3.get(value, 0) + 1

            # проверяю соответсвие правилу по блокам
            for key, value in d3.items():
                if value != 1 and key != '.':
                    return False

    return True
```
