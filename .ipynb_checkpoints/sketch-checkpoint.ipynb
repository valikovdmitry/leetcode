```python
s = "AABABBA"
k = 1
```

```python
def sol(s,k):
    
    l = 0
    freq = [0] * 26   
    maxFreq = 0
    
    maxLen = 0

    for r in range(len(s)):

        while r-l+1 > maxFreq + k:
            idx_remove = ord(s[l]) - ord('A')
            freq[idx_remove] -= 1
            l += 1

        idx_add = ord(s[r]) - ord('A')
        freq[idx_add] += 1
        maxFreq = max(freq[idx_add], maxFreq)

        maxLen = max(r-l+1, maxLen)
        

    return maxLen
```

```python
sol(s,k)
```

```python
s = "AABABBA"
k = 1
```

```python
def sol(s,k):

    l = 0
    freq = [0] * 26
    maxFreq = 0
    maxLen = 0

    for r in range(len(s)):

        # добавляем новый элемент справа
        idx_add = ord(s[r]) - ord('A')
        freq[idx_add] += 1

        # обновляем текущий максимум частоты внутри окна
        maxFreq = max(maxFreq, freq[idx_add])


        # после того как добавили, проверяем осталось ли окно валидным
        # если нет сжимаем его слева
        while r-l+1 > maxFreq + k:
            idx_remove = ord(s[l]) - ord('A')
            freq[idx_remove] -= 1

        # получили в итоге валидное окно, обновили макс длинну
        maxLen = max(maxLen, r-l+1)

    return maxLen

        
```

```python
sol(s,k)
```
