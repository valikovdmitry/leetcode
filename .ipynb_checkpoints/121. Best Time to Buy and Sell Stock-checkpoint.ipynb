```python
prices = [7,1,5,3,6,4]
```

```python
def sol(prices):

    buy, sell = 0, 1
    max_stonks = 0

    while sell < len(prices):
        print(f"--- buy: {buy}, sell: {sell}")
        print(f"--- prices[buy]: {prices[buy]}, prices[sell]: {prices[sell]}")
        

        if prices[buy] > prices[sell]:
            buy = sell
            sell = buy + 1

            print(f"buy: {buy}, sell: {sell}")
            print(f"prices[buy]: {prices[buy]}, prices[sell]: {prices[sell]}")

        else:
            while sell < len(prices) and prices[buy] <= prices[sell]:
    
                stonks = prices[sell] - prices[buy]
                max_stonks = max(stonks, max_stonks)
    
                sell += 1

                

    return max_stonks

      
        
```
