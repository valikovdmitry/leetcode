```python
prices = [7,1,5,3,6,4]
```

```python
def sol(prices):

    l, r = 0, 1
    max_profit = 0

    while r < len(prices):

        if prices[l] <= prices[r]:   
            profit = prices[r] - prices[l]
            max_profit = max(profit, max_profit)
            
        else:
            l = r

        r += 1
    
    return max_profit
```

```python
sol(prices)
```

```python
def sol(prices):

    min_price = prices[0]
    max_profit = 0

    for price in prices:
        if price < min_price:
            min_price = price
        else:
            max_profit = max(max_profit, price - min_price)

    return max_profit
```
