```python
def sol(temperatures):
    
    waiting = []
    res = [0] * len(temperatures)
    
    
    for i, current in enumerate(temperatures):

        while waiting and current > temperatures[waiting[-1]]:

                resolved = waiting.pop()
                res[resolved] = i - resolved

        waiting.append(i)


    return res
```
