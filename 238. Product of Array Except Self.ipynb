```python
def sol(nums):
    prefix = []
    suffix = []
    res = []
    
    for i in nums:
        if len(prefix) == 0:
            prefix.append(i)
        else:
            prefix.append(i*prefix[-1])

    
    for i in range(len(nums) - 1, -1, -1):
        if len(suffix) == 0:
            suffix.append(nums[i])
        else:
            suffix.append(nums[i]*suffix[-1])
            

    for i in range(len(nums)):
        if i == 0:
            res.append(suffix[-2])
        elif i == len(nums) - 1:
            res.append(prefix[-2])
        else:
            res.append(prefix[i-1] * suffix[-i-2])

    return res
```
