```python
def sol(s, k):

    l = 0
    res = 0
    foul = deque()

    for r in range(len(s)):
        print(f'l: {l} r: {r}')
        
        while s[r] != s[l]:
            print(f'--- l: {l} r: {r}')
            if k > 0:
                k -= 1
                print(f'--- --- l: {l} r: {r}')
                break
            else:
                l += 1

        
        
        curr_len = (r - l) + 1
        res = max(curr_len, res)
        r += 1

    return res
```

```python
s = "ABBB"
k = 1
```

```python
def sol(s, k):

    l = 0
    res = 0
    foul = deque()

    for r in range(len(s)):
        print(f'l: {l} r: {r}')
        
        while s[r] != s[l]:
            print(f'--- Wrong ch  --- l: {l} r: {r}')
            if k > 0:
                k -= 1
                foul.appendleft(l)
                print(f'---   Foul    --- l: {l} r: {r} foul: {foul}')
                break
            else:
                if foul:
                    l = foul.pop() 
                    print( f'--- No more k --- : new L: {l} foul: {foul}')
                else:
                    l += 1

        
        
        curr_len = (r - l) + 1
        print(f'curr_len: {curr_len}')
        res = max(curr_len, res)
        r += 1

    print(foul)
    
    return res
```

```python
# как запомнил решение
# прошлая задача была про то, как в паттерне sliding window
# использовать паттер "seen" для проверки на уникальность
# эта же задача про то как использовать частотный счетчик "freq"

```

```python
s = "ABAB"
k = 2
```

```python
def sol(s, k):

    l = 0
    freq = {}
    maxFreq = 0

    maxLen = 0

    for r in range(len(s)):

        # обновляем счетчик текущего символа
        freq[s[r]] = freq.get(s[r], 0) + 1

        # обновляем счетчик масимальной частоты
        maxFreq = max(freq[s[r]], maxFreq)
 

        # проверяем инвариант
        # ЕСЛИ НЕ инвариант ДЕЛАЕМ то, что снизу
        while (r - l) + 1 > maxFreq + k:
            # удаляем из словаря символ 1 раз под левым указателем
            freq[s[l]] -= 1
            # сдвигаем указатель
            l += 1

        # Если мы в инварианте, то мы блок вайл просто пропускаем

        # В любом случае после получения следующего валидного окна
        # обновляем лучшую длину
        maxLen = max((r-l)+1, maxLen)

    return maxLen
    
    
```

```python
sol(s, k) 
```
