```python
nums = [0,3,7,2,5,8,4,6,0,1]

def sol(nums):
    seen = dict()

    for i in nums:
        print(i)
        
        if i not in seen:
            seen[i] = {i}
            print('---', seen[i])
        
        if i-1 in seen:
            seen[i].update(seen[i-1])
            print('---', seen[i])
        
        if i+1 in seen:
            seen[i].update(seen[i+1])
            print('---', seen[i])

        if i-1 in seen:
            seen[i-1].update(seen[i])

        if i+1 in seen:
            seen[i+1].update(seen[i])
        
    
    print()
    
    for k, v in seen.items():
        print(k,v)

    print()
    
    max_len = 0
    for i in seen.values():
        if len(i) > max_len:
            print(i)
            max_len = len(i)

    return max_len
```

```python
nums = [0,3,7,2,100,5,8,4,6,0,1]

def sol2(nums):
    
    s = set(nums)

    print(s)

    c_main = 0
    
    for i in s:
        if i-1 not in s:
            c = 1
            print(i)
            while i+1 in s:
                
                c += 1
                i += 1
                print(i,c)

                if c > c_main:
                    c_main = c

    print(c_main)
            
```

```python
nums = [100,4,200,1,3,2]

sol2(nums)
```

```python
def sol3(nums):
    
    if not nums:
        return 0

    nums_set = set(nums)
    
    max_len = 1
    
    for i in nums_set:
        if i-1 not in nums_set:
            c = 1
            while i+1 in nums_set:
                
                c += 1
                i += 1

                if c > max_len:
                    max_len = c

    return max_len
```

```python
nums = [100,4,200,1,3,2]

sol3(nums)
```
