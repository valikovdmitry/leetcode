```python
ransomNote = "aad"
magazine = "aabcad"
```

```python
def sol(ransomNote, magazine):
    d_magazine = {}

    for i in magazine:
        d_magazine[i] = d_magazine.get(i, 0) + 1

    for i in ransomNote:
        if d_magazine.get(i,0) == 0:
            return False
        d_magazine[i] -= 1

    return True
```

```python
sol(ransomNote, magazine)
```

```python
def sol2(ransomNote, magazine):
    freq = [0] * 26

    for char in magazine:
        idx = ord(char) - ord('a')
        freq[idx] += 1

    for char in ransomNote:
        idx = ord(char) - ord('a')
        if freq[idx] == 0:
            return False
        freq[idx] -= 1

    return True

sol2(ransomNote, magazine)
```
