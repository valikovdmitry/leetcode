```python
def sol(strs):
    
    # создаю хеш таблицу - индекс слова : счетчик букв
    
    dicts = {}

    for idx, word in enumerate(strs):
        d_word = {}
        for char in word:
            d_word[char] = d_word.get(char, 0) + 1
        dicts[idx] = d_word


    # создаю словарь с уникальными счетчиками букв

    groups = []

    for key, value in dicts.items():
        if value not in groups:
            groups.append(value)


    # создаю список со списками, их столько сколько групп
    
    res = []
    for i in groups:
        res.append([])


    # расфасовываю слова по заготовленным слотам
    
    for key, value in dicts.items():

        destination_idx = groups.index(value)
        source = strs[key]
    
        res[destination_idx].append(source)

    return res
        
```

```python
sol(strs)
```
