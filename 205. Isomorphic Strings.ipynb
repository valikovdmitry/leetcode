```python
def sol(s, t):
    s_dict = dict()

    for i, n in enumerate(s):
        if n not in s_dict:
            s_dict[n] = [0, []]
            
        s_dict[n][0] += 1
        s_dict[n][1].append(i)
        

    t_dict = dict()

    for i, n in enumerate(t):
        if n not in t_dict:
            t_dict[n] = [0, []]
            
        t_dict[n][0] += 1
        t_dict[n][1].append(i)

        
    print(s_dict)
    print(t_dict)

    print(list(s_dict.values()))
    print(list(t_dict.values()))
    
    return list(s_dict.values()) == list(t_dict.values())

    
```

```python
s = "bbbaaaba"
t = "aaabbbba"

sol(s, t)
```

```python
def sol2(s, t):
    d = dict()

    for i in range (len(s)):
        
        main_char = s[i]
        check_char = t[i]

        if main_char in d and check_char != d[s[i]]:
            return False
        
        d[s[i]] = t[i]

    return True
```

```python
def sol3(s, t):
    d = dict()
    not_available = set()

    for i in range (len(s)):
        print(s[i], t[i])
        
        
        if s[i] in d and t[i] != d[s[i]]:
            print('false потому что не та пара')
            return False
        
        elif s[i] not in d and t[i] in not_available:
            print('false потому что пара занята другой буквой')
            return False

        elif s[i] not in d and t[i] not in not_available:
            d[s[i]] = t[i]
            not_available.add(t[i])
            print(s[i], '---', d[s[i]])

    return True
```

```python
s = "aab"
t = "aaa"

sol3(s, t)
```

```python
def sol4(s, t):
    d = dict()

    available = [1] * 26 

    for i in range (len(s)):

        idx = ord(t[i]) - ord('a')
        
        if s[i] in d and t[i] != d[s[i]]:
            return False
        
        else:
            if available[idx] == 0:
                return False
            
            d[s[i]] = t[i]
            available[idx] -= 1

    return True
```

```python
s = "badc"
t = "baba"

sol4(s, t)
```

```python
def sol4(s, t):
    map_s = dict()
    map_t = dict()

    for i in range (len(s)):
        print(map_s)
        print(map_t)
        
        if s[i] not in map_s:
            if t[i] not in map_t:
                map_s[s[i]] = [t[i], i]
                map_t[t[i]] = [s[i], i]
            else:
                print('false потому что пара занята')
                return False

        elif s[i] in map_s:
            if map_s[s[i]][0] != t[i]:
                print(map_s[s[i]])
                print(t[i])
                print('false потому что не та пара')
                return False

    return True

```

```python
s = "badc"
t = "baba"

sol4(s, t)
```

```python
def sol5(s, t):
    map_s = dict()
    map_t = set()

    for i in range (len(s)):
        
        if s[i] not in map_s:
            if t[i] not in map_t:
                map_s[s[i]] = t[i]
                map_t.add(t[i])
            else:
                return False

        else:
            if map_s[s[i]] != t[i]:
                return False

    return True
```

```python
s = "abab"
t = "baba"

sol5(s, t)
```
