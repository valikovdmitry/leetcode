```python
def sol2(height):
    
    arr = [[0,0] for i in range(len(height))]

    m = 0
    for i in range(len(height)):
        arr[i][0] = m
        m = max(m, height[i])

    m = 0
    for i in reversed(range(len(height))):
        arr[i][1] = m
        m = max(m, height[i])
        
    print(arr)

    c = 0
    for i in range(len(height)):
        amount = min(arr[i]) - height[i]
        if amount > 0:
            c += amount
        
    return c
```
