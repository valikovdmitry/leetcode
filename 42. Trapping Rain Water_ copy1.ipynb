```python
height = [0,1,0,2,1,0,1,3,2,1,2,1]
```

```python
def sol(height):

    l, r = 0, len(height) - 1
    leftMax, rightMax = height[l], height[r]

    res = 0

    while l <= r:
        print(l, r)
        print('leftMax', leftMax)
        print('rightMax', rightMax)

        if leftMax <= rightMax:
            # считаем количество воды над текщим height[l]
            amount = leftMax - height[l]
            if amount > 0:
                res += amount 

            # обновляем leftMax если он изменился
            leftMax = max(height[l], leftMax)
            
            # двигаем указатель, потому что текущую позицию мы отработали
            l += 1
            
        else:
            # считаем количество воды над текщим height[r]
            amount = rightMax - height[r]
            if amount > 0:
                res += amount

            # обновляем leftMax если он изменился
            rightMax = max(height[r], rightMax)

            # двигаем указатель, потому что текущую позицию мы отработали
            r -= 1

    return res
            
```

```python
def sol2(height):

    l, r = 0, len(height) - 1
    leftMax, rightMax = height[l], height[r]

    res = 0

    while l < r:


        if leftMax < rightMax:
            # текущее место уже "посчитано"
            # сдвигаем указатель l вправо
            l += 1

            # обновляем leftMax, потому что текущая позиция может оказаться новым максимумом
            leftMax = max(leftMax, height[l])

            # вычисляем текущую воду с учетом нового макса
            # меньше 0 быть не может, потому что в крайнем случае leftMax == height[l]
            res += leftMax - height[l]
            
        else:
            # текущее место уже "посчитано"
            # сдвигаем указатель r влево
            r -= 1

            # обновляем rightMax, так как текущая новая высота может стать rightMax
            rightMax = max(rightMax, height[r])

            # считаем количество воды над текущей высотой
            res += rightMax - height[r]
        

    return res
            
```
