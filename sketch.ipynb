```python
nums = [1,3,-1,-3,5,3,6,7]
k = 3
```

```python
from collections import deque

def sol(nums,k):
    q = deque()
    res = []

    for i, n in enumerate(nums):
        
        while q and n > nums[q[-1]]:
            q.pop()
        q.append(i)

        if i + 1 >= k:
            print(i, q)
            
            # дропнуть нелеквид индексы
            if q[0] == i - k:
                q.popleft()

            res.append(nums[q[0]])

    
    return res
```

```python
sol(nums,k)
```
