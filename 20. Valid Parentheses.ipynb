```python
def sol(s):


    stack = []
    rule = {'(':')',
           '[':']',
           '{':'}'}

    for ch in s:
        
        if ch in rule:
            stack.append(rule[ch])
        else:
            # если это закрывающая скобка
            # обшка если:
                # стек пустой
                # не верный символ в конце стека
            if not stack or ch != stack.pop():
                return False

    if not stack:
        return True

    return False

    
    
```
