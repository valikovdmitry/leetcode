```python
def sol(s1, s2):
    
    goalFreq = {}

    for ch in s1:
        goalFreq[ch] = goalFreq.get(ch, 0) + 1

    l = 0
    currFreq = {}
    currLen = 0

    for r in range(len(s2)): 
        # ввожу переменные для текущих букв для читаемости
        ch_l = s2[l]
        ch_r = s2[r]

        # мы проходим первые n символов не трогая левый указатель, пока окно не станет равно длинне паттерна

        # если длина словаря уже равна длинне паттерна, то освобождаем место для правой буквы,
        # путем того, что:
        #    - мы удалеяем левую букву из словаря
        #    - уменьшаем переменную в которой храним длинну окна
        #    - запускаем движение левого указателя
        if currLen == len(s1):
            currFreq[ch_l] -= 1
            currLen -= 1
            l += 1
            

        # когда есть свободное место мы добавляем правый символ в счетчик частот
        # так же увеличиваем переменную в которой храним длинну окна
        currFreq[ch_r] = currFreq.get(ch_r, 0) + 1
        currLen += 1

        # в итоге имея валидное окно с актуальным счетчиком частот, мы сравниваем его с целевым
        # если они равны, то прерываем цикл и возвращаем результат
        if currFreq == goalFreq:
            return True


    return False
```

```python
def sol(s1, s2):

    n, m = len(s1), len(s2)

    if n > m:
        
        return False
    
    goal = {}
    for i in range(n):
        goal[s1[i]] = goal.get(s1[i], 0) + 1

    window = {}
    for i in range(n):
        window[s2[i]] = window.get(s2[i], 0) + 1

    if window == goal:
        return True

    l = 0
    for r in range(n, m):
        
        # убрать l из словаря
        window[s2[l]] -= 1
        if window[s2[l]] == 0:
            window.pop(s2[l])
        l += 1

        # добавить r в словарь
        window[s2[r]] = window.get(s2[r], 0) + 1
        r += 1

        # проверить
        print(window, goal)
        if window == goal:
            return True

 
    return False
```
