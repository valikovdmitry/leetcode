```python
target = 12
position = [10,8,0,5,3]
speed    = [ 2,4,1,1,3]

target = 10
position = [6,8]
speed = [3,2]

target = 10
position = [0,4,2]
speed = [2,1,3]
```

```python
def sol(target, position, speed):
    
    zipped = zip(position, speed)
    data = sorted(zipped, key=lambda row: row[0])
    
    heads = []
    for p, s in data:
        time = (target - p) / s

        while heads and time >= heads[-1]:
            heads.pop()

        heads.append(time)
    

    return len(heads)
```

```python
sol(target, position, speed)
```
