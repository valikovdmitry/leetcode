```python
nums = [1,2,3]

def sol(nums):

    nums.sort()

    for i in range(1, len(nums)):
        if nums[i] == nums[i-1]:
            return True
    

    return False
```

```python
sol(nums)
```

```python
def sol2(nums):
    seen = set()

    for n in nums:
        if n in seen:
            return True
        else:
            seen.add(n)
    
    return False
```

```python
sol2(nums)
```
