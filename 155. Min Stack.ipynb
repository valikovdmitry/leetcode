```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.mins = []
        
    
    def push(self, val: int) -> None:
        self.stack.append(val)

        if not self.mins or (self.mins and val <= self.mins[-1]):
            self.mins.append(val)

    
    def pop(self) -> None:
        last = self.stack.pop()
       
        if self.mins[-1] == last:
            self.mins.pop()
        
        return last
        

    def top(self) -> int:
        return self.stack[-1]
        

    def getMin(self) -> int:
        return self.mins[-1]
        


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

```python
val = 5

obj = MinStack()
obj.push(-2)
print('stack:', obj.stack)
print(' mins:', obj.mins)
obj.push(0)
print('stack:', obj.stack)
print(' mins:', obj.mins)
obj.push(-3)
print('stack:', obj.stack)
print(' mins:', obj.mins)
obj.pop()
print('stack:', obj.stack)
print(' mins:', obj.mins)
param_3 = obj.top()
print(param_3)
param_4 = obj.getMin()
print(param_4)
```
