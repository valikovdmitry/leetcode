```python
pattern = "abba"
s = "dog cat cat fish"
```

```python
def sol(pattern, s):
    s_list = s.split(' ')

    if len(pattern) != len(s_list):
        return False
    
    map_char = dict()
    map_word = set()

    for i in range(len(pattern)):
        
        if pattern[i] not in map_char:
            if s_list[i] in map_word:
                print('false потому пара занята')
                return False
            map_char[pattern[i]] = s_list[i]
            map_word.add(s_list[i])

        if map_char[pattern[i]] != s_list[i]:
            print('false потому что не та пара')
            return False

    return True
            
                
    
```

```python
sol(pattern, s)
```
