```python
tokens = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
```

```python
def sol(tokens):

    stack = []

    for n in tokens:
        print(n)


        if len(n) == 1 and not n.isnumeric():
            print('sym', n, stack)

            b, a = int(stack.pop()), int(stack.pop())
            print(a,b)

            if n == '+':
                stack.append(a + b)
            elif n == '-':
                stack.append(a - b)
            elif n == '*':
                stack.append(a * b)
            elif n == '/':
                stack.append(a / b)
        
        else:
            stack.append(n)
            print('num', n, stack)
            
            
    return stack
```

```python
def sol(tokens):

    stack = []

    for n in tokens:

        if len(n) == 1 and not n.isnumeric():

            b, a = int(stack.pop()), int(stack.pop())

            if n == '+':
                stack.append(a + b)
            elif n == '-':
                stack.append(a - b)
            elif n == '*':
                stack.append(a * b)
            elif n == '/':
                stack.append(a / b)
        
        else:
            stack.append(n)
            
            
    return stack[-1]
```
