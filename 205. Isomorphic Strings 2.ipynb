```python
s = "bbbaaaba"
t = "aaabbbba"
```

```python
def sol(s, t):
    s_lst = [0] * 128
    t_lst = [0] * 128

    for i in s:
        idx = ord(i) - ord('a')
        s_lst[idx] += 1

    for i in t:
        idx = ord(i) - ord('a')
        t_lst[idx] += 1

    print(sorted(s_lst))
    print(sorted(t_lst))
    
    return sorted(s_lst) == sorted(t_lst)

    
```

```python
sol(s,t)
```
