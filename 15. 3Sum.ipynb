```python
nums = [0,0,0,0]
```

```python
def sol(nums):
    res = []
    nums.sort()

    for i in range(len(nums)):

        if nums[i] == nums[i-1] and i != 0:
            continue
        
        left = i + 1
        right = len(nums) - 1
    
        while left < right:
            target = - nums[i]
    
            curr = nums[left] + nums[right]

            
            if curr ==  target:
                res.append([nums[i], nums[left], nums[right]])

            
            if curr < target:

                while left < right and nums[left] == nums[left + 1]:
                    left += 1
            
            else:
                
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1

    return res
```

```python
sol(nums)
```
