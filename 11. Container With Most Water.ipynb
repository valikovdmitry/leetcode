```python
def sol(height):

    left = 0
    right = len(height) - 1

    max_area = 0

    while left < right:
        
        min_value = min(height[left], height[right])
        area = min_value * ( right - left )
        
        print(height[left], height[right], area)
        
        if area > max_area:
            max_area = area

        # почему двигаем меньший:
        # ширина всегда падает, нам нужно хоть как-то 
        # увеличивать площадь, через возможно большую высоту
        # 
        # тут еще понятие ограничивающей высоты
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1


    return max_area
```

```python
def sol2(height):

    left = 0
    right = len(height) - 1

    max_area = 0

    while left < right:
        
        h = min(height[left], height[right])
        width = right - left
        area = h * width
        
        max_area = max(max_area, area)

        if height[left] < height[right]:
            left += 1
        else:
            right -= 1


    return max_area
```

```python
def maxArea(self, height: List[int]) -> int:
    left = 0
    right = len(height) - 1
    max_area = 0

    while left < right:
        # текущая высота контейнера ограничена более низкой стенкой
        h = min(height[left], height[right])
        width = right - left
        max_area = max(max_area, h * width)

        # двигаем более низкую стенку внутрь
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return max_area
```
