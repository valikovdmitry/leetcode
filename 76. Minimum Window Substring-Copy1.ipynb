```python
def sol(s,t):

    need = {}
    for ch in t:
        need[ch] = need.get(ch, 0) + 1

    window = {}
    formed = 0
    required = len(set(t))
    l = 0
    word = s

    flag = False


    for r in range(len(s)):   
        
        window[s[r]] = window.get(s[r], 0) + 1
        
        if s[r] in need and window[s[r]] == need[s[r]]:
            formed += 1

        while formed == required:
            flag = True
            
            if s[l] in need:
                window[s[l]] -= 1

                if window[s[l]] < need[s[l]]:
                    formed -= 1

            if len(s[l:r+1]) < len(word):
                word = s[l:r+1]
            
            l += 1

        
    if not flag:
        return ''
        
    return word
```
