```python
s = "loveleetcode"

def sol(s):
    d = {}
    
    for i in s:
        d[i] = d.get(i, 0) + 1

    print(d)
    
    for i in range(len(s)):
        if d[s[i]] == 1:
            return i

    return -1
    
    

sol(s)
```

```python
def  sol2(s):
    lst = [0] * 26

    for i in s:
        idx = ord(i) - ord('a')
        lst[idx] += 1

    for i, n in enumerate(s):
        idx = ord(n) - ord('a')
        if lst[idx] == 1:
            return i
    return -1
        
sol2(s)
```

```python
def sol3(s):
    tested = set()
    result = []

    for i in s:
        if i in tested:
            continue
        if i not in result:
            result.append(i)
        elif i in result:
            tested.add(i)
            result.remove(i)

    return s.index(result[0])

    print(result)
    print(tested)
    
sol3(s)
```
